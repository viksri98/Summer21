### Exercise 1: Create a function `draw_square()`
`draw_square(xcoord, ycoord, side_length, fill_color, fill_flag)`   
where:  
`xcoord` = the x coordinate location of the bottom left corner  
`ycoord` = the y coordinate location of the bottom left corner  
`side_length` = integer length of a side  
`fill_color` = pick the name of the color from below  
![title](https://cs111.wellesley.edu/content/labs/lab01//images/colorsPage1.png)  
`fill_flag` = True or False (gives you the option of filling in the square or not)

Reproduce the image from Example 1 using the function.

```python
import turtle

t = turtle.Pen()
t.speed(10)
turtle.screensize(canvwidth=600, canvheight=600)


def draw_square(xcoord, ycoord, side_length, fill_color, fill_flag):
    t.home()
    t.penup()
    t.goto(xcoord, ycoord)
    t.fillcolor(fill_color)
    t.begin_fill()
    t.forward(side_length)
    t.left(90)
    t.forward(side_length)
    t.left(90)
    t.forward(side_length)
    t.left(90)
    t.forward(side_length)
    t.end_fill()
    
draw_square(0,0,100, 'blue', True)
draw_square(-150,0,100,'green',True)
draw_square(-150,200,100,'red',True)
draw_square(200,-150,100,'black', True)
turtle.done()
```
