### Exercise 1: Quadratic functions $y = a(x-h)^2 + k$

Write the code to get a plot as follows:

HINT: <br>
`a = np.arange(-10,11, dtype='int')
 h = 0
 k = 0`

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.arange(-10, 11, dtype='int')
x = np.linspace(-10, 10, 20)
h = 0
k = 0

for index in a:
        y = a[index]*((x-h)**2)+k
        plt.plot(x, y)      
plt.title("Quadratic Function")           #Adds a title to the plot


plt.xlabel('x values') # Adds an x axis label
plt.ylabel('y values') # Adds a y axis label
plt.grid()
plt.show()
```

### Exercise 4: Cubic Functions $y=a(x-h)^3 + k$

Graph the three outputs as subplots as shown below. 




HINT: Cut and paste from the previous three exercises. Also refer to matplotlib.ipynb for subplots. 

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.arange(-10, 11, dtype='int')
x = np.linspace(-2, 2, 20)
h = 0
k = 0

plt.subplot(131)
for index in a:
        y = a[index]*((x-h)**3)+k
        plt.plot(x, y)      
plt.title("Slope Change")           #Adds a title to the plot
plt.xlabel('x values') # Adds an x axis label
plt.ylabel('y values') # Adds a y axis label
plt.grid()

a = 2
x = np.linspace(-10, 10, 20)
h = np.arange(-5,6, dtype='int')
k = 0

plt.subplot(132)
for index in h:
    y = a*(x-h[index])**3+k
    plt.plot(x, y)      
plt.title("Horizontal shift")           #Adds a title to the plot
plt.xlabel('x values') # Adds an x axis label
plt.ylabel('y values') # Adds a y axis label
plt.grid()

a = 2
x = np.linspace(-2, 2, 20)
k = np.arange(-5,6, dtype='int')
h = 0

plt.subplot(133)
for index in k:
    y = a*(x-h)**3+k[index]
    plt.plot(x, y)      
plt.title("Vertical shift")           #Adds a title to the plot
plt.xlabel('x values') # Adds an x axis label
plt.ylabel('y values') # Adds a y axis label
plt.grid()

plt.suptitle("Cubic Function")
plt.show()
```

### Exercise 6: Exponential Function y = a * 2<sup>(x-h)</sup> + k
