# Exercise 2: 

**Linear function:**
* $y = a(x-h) + k = mx + b$

**Quadratic function:**
* $y=a(x-h)^2 +k$ 

**Cubic function:**
* $y=a(x-h)^3 + k$ 


Ask the user to select between a linear, quadratic or cubic function using an `input` command.

Get a, h and k using an `input` command (look at Review_Lessons1_7)

HINT:

Use `if`, `elif`, `else` to select among the functions.  
 
Expected Output:<br>
What would you like to plot? L for linear, Q for quadratic, or C for cubic:Q<br>
What is the value of a?1<br>
What is the value of h?0<br>
What is the value of k?0<br>


```python
func_type = input("What would you like to plot? L for linear, Q for quadratic, or C for cubic:")
a = int(input("what is a?"))
h = int(input("what is h?"))
k = int(input("what is k?"))
x_values = np.linspace(-10, 10, 100)

if func_type == 'L':
    y = a * (x_values - h) + k
elif func_type == 'Q':
    y = a * (x_values - h) ** 2 + k
else:
    y = a * (x_values - h) ** 3 + k
plt.plot(x_values, y)
plt.show()
```
