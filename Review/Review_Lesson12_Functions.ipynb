### Exercise 1: Quadratic functions $y = a(x-h)^2 + k$

Write the code to get a plot as follows:

HINT: <br>
`a = np.arange(-10,11, dtype='int')
 h = 0
 k = 0`

```python
import numpy as np
from matplotlib import pyplot as plt

a = np.arange(-10, 11, dtype='int')
x = np.linspace(-10, 10, 20)
h = 0
k = 0

# Your code goes here
for i in a:
    plt.plot(x, i * (x - h) ** 2 + k)
plt.show()
```
