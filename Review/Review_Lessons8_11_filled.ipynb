### Exercise 1: 
Plot a Radical function:
$y= \frac{1}{x}$
Use Exercise 1 in matplotlib.ipynb to help you graph this function. 

NOTE: 1/0 is a math error. **x value cannot be 0.**<br>
HINT: x values should range from -1 to 1. Break it up into two arrays `[-1 to 0] and [0 to 1]`<br>

```python
import numpy as np
from matplotlib import pyplot as plt

xp_values = np.linspace(0.01, 1, 100) #100 evenly spaced values from 0 to 5
xn_values = np.linspace(-1, -0.01, 100) #100 evenly spaced values from -5 to -0
y1 = 1/xn_values
y2 = 1/xp_values
plt.plot(xn_values, y1, label='Radical function', color='c')      #Added in legend
plt.plot(xp_values, y2, label='Radical function', color='c')          #Added in legend
plt.xlabel('x values')                   # Adds an x axis label
plt.ylabel('y values') # Adds a y axis label
plt.legend()                                              #shows the legend
plt.grid()
plt.title("Radical Function")           #Adds a title to the plot
plt.show()
```

### Exercise 2: 

**Quadratic function:**
* $y=a(x-h)^2 +k$ 

**Cubic function:**
* $y=a(x-h)^3 + k$ 

Ask the user to select between a quadratic and cubic function. If the user does not choose a quadratic or cubic function then default to a linear function.<br>

**Linear function:**
* $y = a(x-h) + k$

Get a, h and k using an `input` command (look at Review_Lessons1_7)

HINT:

Use `if`, `elif`, `else` to select among the functions.  
 
Expected Output:<br>
What would you like to plot? Q for quadratic of C for cubic:Q<br>
What is the value of a?1<br>
What is the value of h?0<br>
What is the value of k?0<br>


```python
import numpy as np
from matplotlib import pyplot as plt

choice = input("What would you like to plot? Q for quadratic of C for cubic:")
a = int(input("What is the value of a?"))
x = np.linspace(-10, 10, 20)
h = int(input("What is the value of h?"))
k = int(input("What is the value of k?"))

if (choice =="Q"):
    y = a*(x-h)**2+k
    plt.plot(x, y, label='Quadratic function', color='c')      #Added in legend
    plt.title("Quadratic Function")           #Adds a title to the plot
elif (choice == "C"):
    y = a*(x-h)**3 + k
    plt.plot(x, y, label='Cubic function', color='b')      #Added in legend
    plt.title("Cubic Function")           #Adds a title to the plot
else:
    y = x
    plt.plot(x, y, label='Linear function', color='g')      #Added in legend
    plt.title("Default Linear Function")           #Adds a title to the plot


plt.xlabel('x values')                   # Adds an x axis label
plt.ylabel('y values') # Adds a y axis label
plt.legend()                                              #shows the legend
plt.grid()
plt.show()
```
