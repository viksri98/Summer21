# Lesson 18 Image Processing with Matplotlib

This is a quick overview of some basic image processing that you can perform with matplotlib.  
For more documentation and indepth understanding of the subject refer to :  
https://matplotlib.org/stable/tutorials/introductory/images.html  

The first thing we are going to do is load an image and display it. 

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

img = mpimg.imread('/home/vikram/Documents/Summer21/src/Faces/Face1.jpg')
arr = np.array(img)
print (arr)
```

```python
# Now let us see what this image looks like by plotting it
imgplot = plt.imshow(arr)
```

```python
#You can draw a square or rectangle on top of your image.

# edit part of the array to have only 0s
# y values first, then x values
arr[100:150, 80:170, :] = np.array([0, 0, 255])
# Display the image
plt.imshow(arr)

plt.show()
```
