This is a quick overview of some basic image processing that you can perform with matplotlib.  
For more documentation and indepth understanding of the subject refer to :

The first thing we are going to do is load an image and display it. 

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img = mpimg.imread('../src/Faces/Face1.jpg')
print (img)
```

```python
# Now let us see what this image looks like by plotting it
imgplot = plt.imshow(img)
```

```python
import matplotlib.patches as patches
from PIL import Image

#You can draw a square or another 'patch' on top of your image.
# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(img)

# Create a Rectangle patch
rect = patches.Rectangle((50, 100), 40, 30, linewidth=1, edgecolor='r', facecolor='black')

# Add the patch to the Axes
ax.add_patch(rect)

plt.show()
```
